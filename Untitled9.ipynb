```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
heart_dataset = pd.read_csv("heart.csv")
heart_dataset.head()
```

```
x=heart_dataset.iloc[:, 0:13].values
y=heart_dataset.iloc[:,13].values
```

```
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```
classifier = Sequential()
classifier.add(Dense(units = 7, kernel_initializer = 'uniform', activation = 'relu', input_dim = 13))
classifier.add(Dropout(p = 0.1))
```

```
classifier.add(Dense(units = 7, kernel_initializer = 'uniform', activation = 'relu'))
classifier.add(Dropout(p = 0.1))
```

```
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```
classifier.fit(x_train, y_train, batch_size = 10, epochs = 1000)
```

```
y_pred = classifier.predict(x_test)
y_pred = (y_pred > 0.5)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```
new_prediction = classifier.predict(sc.transform(np.array([[85, 1, 0, 148, 203, 0, 1, 161, 0, 0, 2, 1, 3]])))
print(new_prediction)
```

```
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
def build_classifier():
  classifier = Sequential()
  classifier.add(Dense(units = 7, kernel_initializer = 'uniform', activation = 'relu', input_dim = 13))
  classifier.add(Dropout(p = 0.1))
  classifier.add(Dense(units = 7, kernel_initializer = 'uniform', activation = 'relu'))
  classifier.add(Dropout(p = 0.1))
  classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
  classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
  return classifier
classifier = KerasClassifier(build_fn = build_classifier, batch_size = 10, epochs = 100)
accuracies = cross_val_score(estimator = classifier, X = x_train, y = y_train, cv = 10)
  
  
```
